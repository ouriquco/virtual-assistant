```python
# Set environment variables and keys

import os

os.environ['AMADEUS_CLIENT_ID'] = 'YOUR_AMADEUS_CLIENT_ID'
os.environ['AMADEUS_CLIENT_SECRET'] = 'YOUR_AMADEUS_CLIENT_SECRET'
os.environ['OPENROUTER_API_KEY'] = 'YOUR_OPENROUTER_API_KEY'
os.environ['HOST'] = 'YOUR_HOST'
os.environ['USER'] = 'YOUR_USER'
os.environ['PASSWORD'] = 'YOUR_PASSWORD'
os.environ['DATABASE'] = 'YOUR_DATABASE'
os.environ['DEBUG'] = 'True'

env_keys = [
    'AMADEUS_CLIENT_ID',
    'AMADEUS_CLIENT_SECRET',
    'OPENROUTER_API_KEY',
    'HOST',
    'USER',
    'PASSWORD',
    'DATABASE',
    'DEBUG'
]

with open('./virtual-assistant/.env', 'w') as f:
    for key in env_keys:
        value = os.environ.get(key, '')
        f.write(f"{key}='{value}'\n")
```
