```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from router import Router
from dotenv import load_dotenv
import os 

load_dotenv()

llm = ChatOllama(
    model="llama3",
    temperature=0.2 
)

os.environ['USER'] = 'root'

system_prompt = ("""Answer the question based on the provided guidelines.
**Guidelines:**
- For questions regarding specific flight details, answer strictly with "flight api"
- For questions about the amount of air traffic that has occured in the past, answer strictly with "database"
- For all other questions, answer strictly with "general"
"""
)

sql_template = '''You are an expert SQL assistant.

Given the following database schema:
{table_info}

Task:
- Write a syntactically correct SQL query for the question below.
- Limit the results to {top_k} rows.
- If the query returns results, use them to answer the question in plain English.
- If the query returns no results, reply: "I don't know."

Question:
{input}

First, output the SQL query (on a single line).  
Then, output the answer (or "I don't know").'''

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)

number_of_results = 5

new_router = Router(
    llm=llm,
    system_prompt=system_prompt,
    prompt_template=prompt,
    number_of_results=number_of_results,
    sql_template=sql_template,
)

# user_query = "Find me a flight from Los Angelas to New York for May 5, 2025 for 1 adult"
# user_query = "What is the weather like in New York?"
user_query = "What is the busiest month for air traffic in the US in 2017?"

response = new_router.route(user_query)
print(response)
```
